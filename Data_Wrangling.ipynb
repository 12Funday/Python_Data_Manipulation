## Filter

### Normal Filter

```python
# pilih kolom yang berisi kata "delay"

flight.filter(like='Delay', axis=1).head()
```

### Filter with defined list

```python
# Kita buat list airport dengan 4 delay tertinggi

df = (
    flight
    [['Origin_Airport', 'Dep_Delay']]
    .groupby(['Origin_Airport'])
    .sum()
    .sort_values(by='Dep_Delay', ascending=False)
    .reset_index()
    .head(4)
)
```

```python
df
```

```python
# Membuat list dari hasil filter

my_list = df['Origin_Airport'].tolist()
my_list
```

```python
# Filter menggunakan list dan tampilkan statistis ringkas

df = flight[(flight['Origin_Airport'].isin(my_list)) & (flight['DayOfWeek']>=5)]
df = df[['Origin_Airport', 'Distance']]
df = df.groupby(['Origin_Airport']).agg(['mean', 'median', 'std', 'min', 'max'])
df
```

```python
# Data merupakan statistik dari 4 airport yang memiliki delay tertinggi

# Untuk tampilan yang lebih baik

df.columns = df.columns.droplevel()
df = df.reset_index()
df
```

### Exclude some rows with a defined list

```python
# Coba lihat list berikut

df = pd.DataFrame({
    'Name':[
            'Maria', 'Marc', 'Julia', 'Mike', 'Sarah', 'Sven', 'Mel', 'Alex', 'John', 'Marlene'
    ],
    'Favorite_Fruit': [
                       'Banana', 'Apple', 'Melon', 'Peach', 'Grape', 'Melon', 'Orange', 'Banana', 'Melon', 'Apple'
    ]
})

df
```

```python
# Select specific rows

value_list = ['Melon', 'Apple']

boolean_value_list = df.Favorite_Fruit.isin(value_list)
filtered_df = df[boolean_value_list]
filtered_df
```

```python
# Exclude specific rows

value_list = ['Melon', 'Apple']
inverse_boolean_list = ~df.Favorite_Fruit.isin(value_list)
inverse_filtered_df = df[inverse_boolean_list]
inverse_filtered_df
```
